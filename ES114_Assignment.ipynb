```python
# Question 1

list_of_numbers =[]
for i in range (1 , 21):
    list_of_numbers.append(i)
print(list_of_numbers)
total = 0
for i in range (9,len(list_of_numbers)) :
    if list_of_numbers[i] % 2 == 0 :
        n = list_of_numbers[i]
        n=list_of_numbers[i]
        total += n
print(total)

import statistics as st
 
mean = st.mean(list_of_numbers)
print("Mean :", mean)
median = st.median(list_of_numbers)
print("median :", median)
standered_deviation = st.stdev(list_of_numbers)
print("standered_deviantion :",standered_deviation)
```

```python
#Question 3

n = int(input())
number = []
for i in range(1,n) :
    number.append(i)
print (number)
list_fibonacci = []
a = 0
b = 1
i = 1 
while i < n :
    m = a + b
    a = b
    b = m
    i += 1
    list_fibonacci.append(m)
print(list_fibonacci)


from matplotlib import pyplot as plt 

x = number 
y = list_fibonacci

plt.plot(x,y)
plt.show()

```

```python
#Question 7



student_marks = {
    "a" : 45,
    "b" : 36, 
    "c" : 66,
    "d" : 54
}

alpha = []
for value in student_marks.values():
    alpha.append(int(value))

i=0
sum=0
while i <= (len(alpha)-1) :
    sum = sum + alpha[i]
    i+=1
average = sum/(len(alpha))
print(average)

for i in alpha :
    if i>average :
        print(i)

beta =[]
for value in student_marks:
    beta.append(value)

print(beta)

from matplotlib import pyplot as pt 

x=beta
y= alpha
pt.bar(x,y)
pt.show
```

```python
#Question 8
n=int(input())

b=[]
number = []
square = []

i=1
while  i <= n :
    c = i
    d = i**2
    e = (c,d)
    b.append(e)
    number.append(i)
    square.append(i**2)

    i+=1

print(b)

from matplotlib import pyplot as plt

x= number 
y= square 
 
plt.scatter(x,y)
plt.show()
```
